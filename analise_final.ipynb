```python
import pandas as pd

arquivo = pd.read_csv("Dados.csv")
#arquivo.info()
display(arquivo.head(25))
```

Calculando quantas vagas estão disponivel por curso em todo o brasil.

```python
qtd_cursos = arquivo[["NOME_CURSO", "REGIAO", "QT_VAGAS_AUTORIZADAS"]].groupby(["NOME_CURSO", "REGIAO"]).sum()

final = qtd_cursos.query('QT_VAGAS_AUTORIZADAS != 0')

display(final)
```

Limitando as informações apenas para as vagas na regiao sudeste do pais.

```python
df = pd.DataFrame(arquivo)

regiao = df[df['REGIAO'] == 'SUDESTE']

cursos = regiao[["NOME_CURSO", "REGIAO", "QT_VAGAS_AUTORIZADAS"]].groupby(["NOME_CURSO", "REGIAO"]).sum()

ativos = cursos.query('QT_VAGAS_AUTORIZADAS != 0')

display(ativos)
```

Limitando as informações para maior cidade da região sudeste.

```python
df = pd.DataFrame(arquivo)

estado = df[df['UF'] == 'SP']

cursos_sp = estado[["NOME_CURSO", "UF", "QT_VAGAS_AUTORIZADAS"]].groupby(["NOME_CURSO", "UF"]).sum()

ativos_sp = cursos_sp.query('QT_VAGAS_AUTORIZADAS != 0')

display(ativos_sp)
```

Mostrando as faculdades de São Paulo e suas quantidades de vagas, sendo são paulo o maior municipio de SP

```python
df = pd.DataFrame(arquivo)

faculdades = df[df['MUNICIPIO'] == 'São Paulo']

faculdades_sp = faculdades[["NOME_IES", "MUNICIPIO", "QT_VAGAS_AUTORIZADAS"]].groupby(["NOME_IES", "MUNICIPIO"]).sum()

ativos_faculdades_sp = faculdades_sp.query('QT_VAGAS_AUTORIZADAS != 0')

display(ativos_faculdades_sp)
```

Mostrando o menor municipio de SP e suas respectivas vagas, sendo que há 1.487 habitantes nessa região.

```python
df = pd.DataFrame(arquivo)

faculdades = df[df['MUNICIPIO'] == 'Flora Rica']

faculdades_sp = faculdades[["NOME_IES", "MUNICIPIO", "QT_VAGAS_AUTORIZADAS"]].groupby(["NOME_IES", "MUNICIPIO"]).sum()

ativos_faculdades_msp = faculdades_sp.query('QT_VAGAS_AUTORIZADAS != 0')

display(ativos_faculdades_msp)
```

Verificando qual a quantidade de vagas liberadas por Grau.

```python
df = pd.DataFrame(arquivo)

faculdades = df[df['MUNICIPIO'] == 'São Paulo']

faculdades_sp = faculdades[["MUNICIPIO", "GRAU", "QT_VAGAS_AUTORIZADAS",]].groupby(["MUNICIPIO","GRAU"]).sum()

grau_faculdades_sp = faculdades_sp.query('QT_VAGAS_AUTORIZADAS != 0')

display(grau_faculdades_sp)
```

Verificando quantidade de vagas por Modalidade.

```python
df = pd.DataFrame(arquivo)

faculdades = df[df['MUNICIPIO'] == 'São Paulo']

faculdades_sp = faculdades[["MUNICIPIO", "MODALIDADE", "QT_VAGAS_AUTORIZADAS",]].groupby(["MUNICIPIO","MODALIDADE"]).sum()

mod_faculdades_sp = faculdades_sp.query('QT_VAGAS_AUTORIZADAS != 0')

display(mod_faculdades_sp)
```

Verificando quantidade de universidade para fins lucrativos.

```python
df = pd.DataFrame(arquivo)

faculdades = df[df['MUNICIPIO'] == 'São Paulo']

faculdades_sp = faculdades[["MUNICIPIO", "CATEGORIA_ADMINISTRATIVA", "QT_VAGAS_AUTORIZADAS",]].groupby(["MUNICIPIO","CATEGORIA_ADMINISTRATIVA"]).sum()

fins_faculdades_sp = faculdades_sp.query('QT_VAGAS_AUTORIZADAS != 0')

display(fins_faculdades_sp)
```
