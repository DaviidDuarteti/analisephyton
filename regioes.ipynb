```python
import pandas as pd

arquivo = pd.read_csv("Dados.csv")
#arquivo.info()
display(arquivo.head(2))
```

Separando as vagas por regioes
Primeira SUDESTE

```python
df = pd.DataFrame(arquivo)

regiao = df[df['REGIAO'] == 'SUDESTE']

cursos = regiao[["NOME_CURSO", "REGIAO", "QT_VAGAS_AUTORIZADAS"]].groupby(["NOME_CURSO", "REGIAO"]).sum()

ativos_sudeste = cursos.query('QT_VAGAS_AUTORIZADAS != 0')

display(ativos_sudeste)

qtsudeste = ativos_sudeste["QT_VAGAS_AUTORIZADAS"].sum()

display(qtsudeste)
```

Criando nova tabela com os dados analisados:

```python
dados = {
    'Regiao': [ 'Sul', 'Sudeste', 'Norte', 'Nordeste', 'Centro_Oeste'],
    'Qt_vagas': [944563809, 2333028227, 637651691, 1218963125, 620100529  ]
}

vagas_regioes = pd.DataFrame(dados)

print(vagas_regioes)

```

Criando grafico das vagas por regioes:

```python
import matplotlib.pyplot as plt



regioes = vagas_regioes["Regiao"]   
qt = vagas_regioes["Qt_vagas"]


plt.figure(figsize=(12, 8))
plt.title("Distribuição de vagas nas regiões do Brasil")
plt.pie(qt, labels=regioes, autopct="%1.2f%%")
plt.show()
```
